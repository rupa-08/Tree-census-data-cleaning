```python
tree_census = pd.read_csv('2015_Street_Tree_Census_-_Tree_Data_20250309.csv')
tree_census
```

```python
#to read all the columns
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows',None)
tree_census
```

```python
#some of the columns are not understandable so the pdf about the dataset helps us to understand better about the dataset.
#listing out all the columns
tree_census.columns
```

```python
# subsetting data with necessary column and removing unnecessary columns like lacation

tree_census_subset = tree_census[['tree_id','tree_dbh', 'stump_diam',
       'curb_loc', 'status', 'health', 'spc_latin', 'steward','sidewalk','problems', 'root_stone',
       'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
       'brch_light', 'brch_shoe', 'brch_other']]
tree_census_subset
```

```python
#checking null values

tree_census_subset.isna().sum()
```

```python
# missing health values
tree_census_subset[tree_census_subset['health'].isna()]
```

```python
#Finding from above output
#from above output we can see that three column are integer values
#the mean tree diameter is 11.27 but the maximum value see to high(450 inch) which seems to be an outlier. We can know that diameter is measured in inch from the pdf description about the data. 
#so is for the stum diameter, where mean is 0.43, std is 3.2 and max is 140. So this also may have outlier.
#tree_id should be a cateforical value but here it is numeric value
```
